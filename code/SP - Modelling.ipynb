```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Bidirectional, LSTM, TimeDistributed
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("preprocessed_dataset.csv")
```

```python
df.info()
```

```python
refined_df = df.drop(columns = ["Unnamed: 0","appeared", "imports"])
y = refined_df["label"]
X = refined_df.drop(columns =["sha256","md5","label"])
y = pd.get_dummies(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
epochs = 10
batch_size = 32
```

```python
# Input layer - Shape matches your 10-feature input 
input_layer = Input(shape=(None, 10))  # None for variable timesteps

# Bidirectional LSTM layer
x = Bidirectional(LSTM(64, return_sequences=True))(input_layer)  

# Additional hidden LSTM layer (optional)
x = Bidirectional(LSTM(32))(x)  

# Optional: TimeDistributed for classification at each timestep
# x = TimeDistributed(Dense(64, activation='relu'))(x)

# Output layer for single classification (3 possible classes)
output_layer = Dense(3, activation='softmax')(x)

# Create the model
model = tf.keras.Model(inputs=input_layer, outputs=output_layer)
```

```python
# Assume 'input_dim' is determined by your chosen input representation
input_layer = Input(shape=(100,))

# Hidden Dense Layers  (Experiment with the number of layers/units)
x = Dense(64, activation='relu')(input_layer)
x = Dense(64, activation='relu')(x)
x = Dense(64, activation='relu')(x)
x = Dense(32, activation='relu')(x) 

# Output Layer for Classification
output_layer = Dense(3, activation='softmax')(x)  # Assuming 3 classes

model = tf.keras.Model(inputs=input_layer, outputs=output_layer)
```

```python
model.compile(optimizer='adam', 
              loss='categorical_crossentropy',  # For multi-class classification
              metrics=['accuracy'])  # Add if needed
```

```python
model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size)
```
